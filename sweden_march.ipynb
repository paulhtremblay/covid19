```python
import pandas as pd
```

```python
from bokeh.plotting  import figure, show
from bokeh.models import Legend
import numpy as np
from bokeh.io import output_notebook
import math

```

```python
def histogram(nums):
    p = figure(title = 'num of countries by deaths')
    hist, edges = np.histogram(nums)
    legen = 'histogram'
    r = p.quad(top = hist, bottom=0, left=edges[:-1], right=edges[1:], alpha = .4) 
    legend = Legend(items=[
            (None , [r]),
            ], location=(10, 0)) 
    p.add_layout(legend, 'right')
    p.xaxis.axis_label = 'deaths/million'
    p.yaxis.axis_label = 'number countrys'

    #legend.background_fill_color = "green"
    return p
```

```python
df = pd.read_csv('~/Downloads/results-20210316-202100.csv')
```

```python
output_notebook()
deaths = df['deaths'].tolist()
deaths = [x for x in deaths if not math.isnan(x)]
deaths = [x for x in deaths if x > 5]
deaths = [x/(1e6/100) for x in deaths]
show(histogram(deaths))
```
