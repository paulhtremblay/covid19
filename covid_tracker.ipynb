```python
def output_graphs(script, div, text, 
                 home_page_dir = '/home/henry/projects/covid19/home_page/'):
    with open(os.path.join(home_page_dir, 'script'), 'w') as write_obj:
              write_obj.write(script)
    with open(os.path.join(home_page_dir, 'div'), 'w') as write_obj:
              write_obj.write(div)
    with open(os.path.join(home_page_dir, 'text.txt'), 'w') as write_obj:
              write_obj.write(text)
```

```python
def make_df():
    df = pd.read_csv('data/covid_tracker_states.csv')
    df['date'] = pd.to_datetime(df['date'])
    return df
```

```python
def make_graph(df, state, plot_width = 300, plot_height = 300, title = None, window = 3):
    df_state = df[df['state']== state]
    dates = df_state['date']
    cases = df_state['new_cases'].rolling(window).mean()
    cases_cov = df_state['positive_increase'].rolling(window).mean()
    tests = df_state['total_test_results_increase'].rolling(window).mean()
    porp = cases/tests
    mean_cases = np.mean(cases)
    mean_tests = np.mean(tests)
    cases_change = cases/mean_cases
    tests_change = tests/mean_tests
    p1 = figure(x_axis_type = 'datetime', title = title, 
                 plot_width = plot_width , plot_height = plot_height, y_range = None)
    p2 = figure(x_axis_type = 'datetime', title = title, 
                 plot_width = plot_width , plot_height = plot_height, y_range = None)
    p3 = figure(x_axis_type = 'datetime', title = title, 
                 plot_width = plot_width , plot_height = plot_height, y_range = None)
    p4 = figure(x_axis_type = 'datetime', title = title, 
                 plot_width = plot_width , plot_height = plot_height, y_range = None)

    p1.line(x = dates, y = cases, color = 'blue', legend_label = 'cases')
    p1.line(x = dates, y = cases_cov, color = 'red', legend_label = 'cases cov')
    p2.line(x = dates, y = tests, color = 'green', legend_label = 'tests')
    p3.line(x = dates, y = porp, color = 'green', legend_label = 'pprp')
    p4.line(x = dates, y = cases_change, color = 'blue', legend_label = 'cases')
    p4.line(x = dates, y = tests_change, color = 'red', legend_label = 'tests')

    p1.legend.location = "top_center"
    p2.legend.location = "top_center"
    p4.legend.location = "top_center"



    return p1, p2, p3, p4

def all_graphs():
    df = make_df()
    data = []
    states = set(df['state'].tolist())
    for state in states:
        p1, p2, p3, p4= make_graph(df, state, window = 7, title = state)
        data.extend([p1,p2, p3, p4])
    grid = gridplot(data, ncols = 4)
    return grid

show(all_graphs())
```
