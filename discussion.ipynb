```python
from bokeh.io import show
from bokeh.plotting import figure
from bokeh.io import output_notebook, reset_output
from bokeh.layouts import gridplot

import pandas as pd
import datetime
import csv
```

```python
df = pd.read_csv('data/seven_day_state.csv')
df['date'] = pd.to_datetime(df['date'])
```

```python
df.head()
df_fl = df[(df['state'] == 'Florida') & (df['date'] >= datetime.datetime(2020, 11, 1))
          & (df['date'] <= datetime.datetime(2021, 1, 31))]
df_fl = df_fl.assign(cases_pop = df_fl['new_cases']/df_fl['population'] * 100000)

df_ca = df[(df['state'] == 'California') & (df['date'] >= datetime.datetime(2020, 11, 1))
          & (df['date'] <= datetime.datetime(2021, 1, 31))]
df_ca = df_ca.assign(cases_pop = df_ca['new_cases']/df_ca['population'] * 100000)
```

```python
output_notebook()
p = figure(x_axis_type = 'datetime', title = 'Covid-19 Cases', 
                 plot_width = 650 , plot_height = 650, y_range = None)
dates = df_fl['date'].tolist()
cases_fl = df_fl['cases_pop']
cases_ca = df_ca['cases_pop'].tolist()
diff = []
for counter, i in enumerate(cases_fl):
    diff.append(i - cases_ca[counter])

p.line(x = dates, y = cases_fl, legend_label = 'Florida')
p.line(x = dates, y = cases_ca, legend_label = 'California', color = 'green')
p.yaxis.axis_label = 'Cases/100,000'


show(p)
```

```python
p = figure(x_axis_type = 'datetime', title = 'Difference: Florida - California', 
                 plot_width = 650 , plot_height = 650, y_range = None)
p.line(x = dates, y = diff, legend_label = 'Difference')
s = [0 for x in dates]
p.line(x = dates, y = s, color = 'red', line_width=1, legend_label = 'no diff')
p.yaxis.axis_label = 'Differences Cases/100,000'

show(p)
```

```python
with open('discussion.csv', 'w') as write_obj:
    csv_writer = csv.writer(write_obj)
    csv_writer.writerow(['date', 'cases_fl', 'cases_ca'])
    for counter, i in enumerate(cases_fl):
        l = []
        l = [dates[counter], i, cases_ca[counter]]
        csv_writer.writerow(l)
```
